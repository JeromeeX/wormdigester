```python
import numpy as np
import matplotlib.pyplot as plt
# Create a multiple bar plot, where the first element in each list of data is in the same category, and second element in each list of data is in the same category, and so on.
data = [[22511, 2821, 4812],
[18063, 17732, 16544]]

X = np.arange(3)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xticks(X + 0.125, ['Control group', '1x SRE', '2x SRE'])
ax.bar(X + 0.00, data[0], color = 'tab:blue', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'tab:green', width = 0.25)
ax.legend(labels=['All reads', 'AC reads'])
# Label the height of the bars
for i in range(len(data)):
    for j in range(len(data[i])):
        plt.text(X[j] + i*0.25, data[i][j], str(data[i][j]), ha='center', va='bottom')
# Add the y-axis label
plt.ylabel('N50 (nt)')
# Add the title
plt.title('N50 of reads')
# Add the x-axis label
plt.xlabel('Groups')

# Automatically adjust subplot parameters to give specified padding
plt.tight_layout()
# Save the figure as a PNG file with a size that is big enough to include all the labels
plt.savefig('fig14_N50comparison.png', dpi=500, bbox_inches='tight')  # Adding bbox_inches='tight' can also help include all elements in the figure

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Existing data
data = [[22511, 2821, 4812], [18063, 17732, 16544]]
# New data to be plotted on secondary y-axis
new_data = [1.89, 3.88, 17.50]

X = np.arange(3)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])  # Primary y-axis

plt.xticks(X + 0.125, ['Undigested', '1x SRE', '2x SRE'])
ax.bar(X + 0.00, data[0], color = 'tab:red', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'tab:blue', width = 0.25)
ax.legend(labels=['All reads', 'AC reads'], loc='upper left')
ax.set_ylabel('N50 (nt)')
ax.set_xlabel('Control and experiment groups')
ax.set_ylim(0, 27000)

# Adding another set of bars with different scale for new_data
# Create a secondary y-axis
ax2 = ax.twinx()
ax2.plot(X + 0.125, new_data, color = 'tab:green', marker='o', linestyle='-', linewidth=2, markersize=8, label='AC enrichment')
ax2.set_ylabel('Enrichment level (%)')
ax2.set_ylim(0, 25)

# Label the height of the bars and points
for i in range(len(data)):
    for j in range(len(data[i])):
        ax.text(X[j] + i*0.25, data[i][j], str(data[i][j]), ha='center', va='bottom')
# Label the height of ax2 data points
for i in range(len(new_data)):
    ax2.text(X[i] + 0.125, new_data[i]+0.2, str(new_data[i]), ha='center', va='bottom')

# Adding legend for the new data
ax2.legend(loc='upper right')

plt.tight_layout()
plt.savefig('fig7.png', dpi=500, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot four datapoints and add a regression line
size = [0.1, 0.106, 5.174, 14.677]
time = [5.5, 8, 123.15, 316.8]
# Calculate the regression line
m, b = np.polyfit(size, time, 1)
plt.scatter(size, time)
# Plot the regression line with dashed lines
plt.plot(size, m*np.array(size) + b, linestyle='dashed', color='tab:red')
# Print the regression line equation and R-squared value on the bottom right corner
plt.text(0.5, 100, 'y = ' + str(round(m, 2)) + 'x + ' + str(round(b, 2)) + '\nR-squared = ' + str(round(np.corrcoef(size, time)[0,1]**2, 2)), color='tab:red')
plt.xlabel('Genome size (Gb)')
plt.ylabel('Time (sec)')
plt.title('Time to process genome data')
plt.savefig('time_vs_size.png', dpi=500, bbox_inches='tight')
plt.show()
```
